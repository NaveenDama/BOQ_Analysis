```python
import pandas as pd
```

```python
df = pd.read_excel("BOQ.xlsx")
df
```

```python
df = df[~df['Amount'].isna()]
```

```python
earth_work_filters = [['earthwork','excavation'],['earth work','excavation'],['filling','excavated'],['filling','gravel']]
```

```python
df['category'] = ['']*len(df)
```

```python
def get_category(description):
#     print(type(description))
    print(description[0][0])

    try:
        description = description.lower()

        for filter_list in earth_work_filters:
#             print(filter_list)
            keyword_len = len(filter_list)
            found_count = 0

            for keyword in filter_list:
                if(keyword in description):
                    found_count += 1
#             print("key found count:",found_count)
            
            if(found_count == keyword_len):
                return "Earth Work"
        return ""
    except:
        print("")
#         print("\n\nException:",description)
            
            
        
```

```python

df.apply(get_category)
```
