```python
import pandas as pd
```

```python
df = pd.read_excel("BOQ.xlsx")
df['category'] = ['not_filled']*len(df)
df = df[~df['Amount'].isna()]
df
```

```python
earth_work_filters = ['Earth Work',['earthwork','excavation'],['earth work','excavation'],
                      ['filling','excavated'],['filling','gravel'],
                      ['filling','sand'],['filling','metal'],['soil']
                     ]
concrete_filters = ['Concrete',['concrete']]
steel_filters = ['Steel',['steel','fe 500']]
brick_masonry_filters = ['Brick Masonry',['brick masonry']]
plastering_filters = ['Plastering',['plastering'],['impervious']]
painting_filters = ['Painting',['painting']]
flooring_filters = ['Flooring',['flooring'],['skirting'],['dadooing'],['granite']]
door_filters = ['Doors',['door']]
window_filters = ['Windows',['window']]
ventilator_filters = ['Ventilators',['ventilator']]
sanitary_filters = ['Sanitary',
                    ['cpvc','pipe'],['soap'],['manhole'],['inspection chamber'],
                    ['swg'],['gully','trap'],['floor','trap'],['orissa','pan'],
                    ['water','closet'],['urinal'],['wash','basin'],['brass','tap'],
                    ['swr','pvc'],['ball','valve','din'],['wall','mixer'],['water','tank']
                   ]
electrical_filters = ['Electrical',['wire'],['wiring'],['fire'],['alaram'],['earthing'],['electrical'],
                      ['electronic'],['cord'],['motor'],
                      ['cable'],['copper'],['frls'],['fan'],['copper cable'],
                      ['switch'],['pvc','frls'],['fan','hook'],['modular','switch'],
                      ['modular','box'],['phase','neutral'],['3 phase'],['distribution','board'],
                      ['mcb'],['single','phase'],['g.i','strip' ],['led','light'],['led','power'],
                      ['batten'],['ding','dong'],['patch'],['net','work'],['network'],['pole','erection'],
                      ['valve'],['hydrant'],['honeywell']
                     ]
```

```python
def apply_filter(description,given_filter):
    category = given_filter[0]
    for filter_list in given_filter[1:]:
#             print(filter_list)
        keyword_len = len(filter_list)
        found_count = 0
        
        for keyword in filter_list:
            if(keyword in description):
                found_count += 1
#             print("key found count:",found_count)

        if(found_count == keyword_len):
            return category
    return "not_filled"
```

```python
def get_category(row,*filters):
#     print(filters)
#     print(type(description))
  
        if(row['category'] == "not_filled"):

                description = row['Description'].lower()
                return apply_filter(description,filters)
        else:
            return row['category']

 
            
            
        
```

```python
df['category'] = df.apply(get_category,axis=1,args=earth_work_filters)
df['category'] = df.apply(get_category,axis=1,args=concrete_filters)
df['category'] = df.apply(get_category,axis=1,args=steel_filters)
df['category'] = df.apply(get_category,axis=1,args=brick_masonry_filters)
df['category'] = df.apply(get_category,axis=1,args=plastering_filters)
df['category'] = df.apply(get_category,axis=1,args=painting_filters)
df['category'] = df.apply(get_category,axis=1,args=flooring_filters)
df['category'] = df.apply(get_category,axis=1,args=door_filters)
df['category'] = df.apply(get_category,axis=1,args=window_filters)
df['category'] = df.apply(get_category,axis=1,args=ventilator_filters)
df['category'] = df.apply(get_category,axis=1,args=electrical_filters)
df['category'] = df.apply(get_category,axis=1,args=sanitary_filters)
df['category'].value_counts()
```

```python
df[df['category'] == 'not_filled'].to_csv("Not filled.csv")
```
