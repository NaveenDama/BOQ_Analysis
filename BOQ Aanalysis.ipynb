```python
def apply_filter(description,given_filter):
    category = given_filter[0]
    for filter_list in given_filter[1:]:
#             print(filter_list)
        keyword_len = len(filter_list)
        found_count = 0
        
        for keyword in filter_list:
            if(keyword in description):
                found_count += 1
#             print("key found count:",found_count)

        if(found_count == keyword_len):
#             print(filter_list)
            return category+"   "+"["+','.join(filter_list)+"]"
    return "not_filled"
```

```python
def get_category(row,*filters):
#     print(filters)
#     print(type(description))
  
        if(row['category_keywords'] == "not_filled"):

                description = row['Description'].lower()
                return apply_filter(description,filters)
        else:
            return row['category_keywords']

 
        
```

```python
def apply_sub_filter(description,filter_dict):
    
    for filter_list in filter_dict.keys():
#         print(filter_list)
        keyword_len = len(filter_list)
        found_count = 0
        
        for keyword in filter_list:
            if(keyword in description):
                found_count += 1
#         print("key found count:",found_count)

        if(found_count == keyword_len):
#             print(filter_dict[filter_list])
            return filter_dict[filter_list]
    return "not_filled"
```

```python
def get_sub_category(row,*filters):

    filter_dict = filters[0]
    category = filters[1]
#     print(filter_dict)
    
#     print(type(description))
  
    if(row['sub_category'] == "not_filled" and row['category'] == category):

            description = row['Description'].lower()
            return apply_sub_filter(description,filter_dict)
    elif(row['sub_category'] == "not_filled" and row['category'] == 'Steel Works'):
        return "Steel Works"
    elif(row['sub_category'] == "not_filled" and row['category'] == 'Wood Works'):
        return "Wood Works"
    else:
        return row['sub_category']
```
